```
import numpy as np
import matplotlib.pyplot as plt
```

```
ntl = lambda x: 10*np.log10(x)
ltn = lambda x: 10**(x/10)

def cos(frek,tipe,ht,hr,d):
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
ht = 80
hr = 1.5
dmin=300 
dmax=1000 
N = 200
pts = 30
gts = 10
grc = 2
bw = 5

d=np.random.randint(dmin,dmax,N)
```

```
def pin(pts,pl,gts,grc):
  pts1 = ntl(pts)
  w = pts - pl + gts + grc
  w1 = ltn(w)
  return w1
#output = w terima
```

```
def snr(pin,bw):
  noise = -174 + 10*np.log10(bw)
  noise1=ltn(noise)
  snr = pin/noise1
  return snr
#output = siignal noise ratio (numerik)
```

```
def ks(snr,bw):
  kps = bw*2*np.log(1+snr)
  return ks
#output = kapasitas seluler
```

```
def pinurban():
  f=1800 
  loc='urban'
  pl=cos(f,loc,ht,hr,d)
  pinfix = pin(pts,pl,gts,grc)
  d_sort=d[np.argsort(d)]
  Urban_sort=pinfix[np.argsort(d)]
  return d_sort,Urban_sort

def pinopen():
  f=1800 
  loc='open'
  pl=cos(f,loc,ht,hr,d)
  pinfix = pin(pts,pl,gts,grc)
  d_sort=d[np.argsort(d)]
  Urban_sort=pinfix[np.argsort(d)]
  return d_sort,Urban_sort

def pinsuburban():
  f=1800 
  loc='suburban'
  pl=cos(f,loc,ht,hr,d)
  pinfix = pin(pts,pl,gts,grc)
  d_sort=d[np.argsort(d)]
  Urban_sort=pinfix[np.argsort(d)]
  return d_sort,Urban_sort
```
